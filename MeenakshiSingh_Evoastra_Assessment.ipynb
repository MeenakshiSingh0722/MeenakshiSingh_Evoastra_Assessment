**Additional Information:**

### Question 4 (15 points)
**Calculate and analyze:**

a) Which customer segment generates the highest **net profit per customer** (considering returns)? Show your calculations. (8 points)

b) Which segment has the **best customer lifetime value** considering churn rates? Provide reasoning. (7 points)


```python
# Your Calculations for Question 4

# a) Net profit per customer calculations:
# Net Profit = (Avg Purchase * Profit Margin) - (Avg Purchase * Profit Margin * Return Rate)

# Young Customers:
# (850 * 0.20) * (1 - 0.12) = 170 * 0.88 = 149.60
# Net Profit: ₹149.60

# Middle-aged Customers:
# (1200 * 0.20) * (1 - 0.08) = 240 * 0.92 = 220.80
# Net Profit: ₹220.80

# Senior Customers:
# (950 * 0.20) * (1 - 0.15) = 190 * 0.85 = 161.50
# Net Profit: ₹161.50

# Result: The Middle-aged segment generates the highest net profit per customer (₹220.80).

# b) Customer Lifetime Value (CLV) Analysis:
# Middle-aged Customers have the best CLV.
# Reasoning: CLV is directly proportional to profit and inversely proportional to churn.
# 1. They have the highest net profit per transaction (₹220.80).
# 2. They have the lowest churn rate (15%) among all segments, meaning they remain customers
#    for a longer duration, significantly increasing their cumulative value over time compared
#    to Seniors (30% churn) or Young customers (25% churn).
```
